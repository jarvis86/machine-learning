```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("/kaggle/input/list-of-mostfollowed-instagram-accounts/List of most-followed Instagram accounts.csv")
```

```python
df.head()
```

```python
df.drop(df.columns[0],axis=1,inplace=True)
```

```python
df.head()
```

```python
df.drop(index=50,inplace=True)
df.drop(index=0,inplace=True)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df.info()
```

```python
df["Followers(millions)[2]"]=df["Followers(millions)[2]"].astype("str").astype("float")
```

```python
plt.figure(figsize=(15,10))
ax=sns.barplot(x=df["Owner"].head(10),
              y=df["Followers(millions)[2]"].head(10),
              data=df,
              palette="Blues")

for i in ax.containers:
    ax.bar_label(i,)
    
plt.title("Top 10 instgram follwers",fontsize=20)
plt.show()
```

```python
plt.figure(figsize=(16,16))
plt.pie(df['Followers(millions)[2]'].head(10),shadow = True, labels=df["Username"].head(10),autopct = "%0.2f%%")
plt.show()
```

```python
plt.figure(figsize=(15,10))
ax=sns.barplot(x=df["Owner"].head(10),
              y=df["Followers(millions)[2]"].head(10),
              data=df,
              palette="Blues")

for i in ax.containers:
    ax.bar_label(i,)
    
plt.title("Top 10 instgram follwers",fontsize=20)
plt.show()
```

```python
plt.figure(figsize=(30,30))
ax1=sns.countplot(x=df["Profession/Activity"].head(10),
              data=df,
              palette="crest")
for i in ax1.containers:
    ax1.bar_label(i,)
    
plt.title("Top 10 instgram follwers by profession",fontsize=20)
plt.show()

```

**Follwers By Country**

```python
plt.figure(figsize=(15,10))
ax2=sns.barplot(x=df["Country/Continent"].head(10),
              y=df["Followers(millions)[2]"].head(10),
              data=df,
              palette="Greens")
for i in ax2.containers:
    ax2.bar_label(i,)
    
plt.title("Top 10 instgram follwers By There Country",fontsize=20)
plt.show()
```

**Most Profession are on Instagram**

```python
plt.figure(figsize=(20, 8))
plt.xticks(rotation=85, fontsize=12) 
ax3=sns.countplot(data=df, x='Profession/Activity',
              order=df["Profession/Activity"].value_counts().sort_values(ascending=False).index)
for i in ax3.containers:
    ax3.bar_label(i,)
    
plt.title("Most Profession are on Instagram",fontsize=20)
plt.show()
```
